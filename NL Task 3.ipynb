# Minimum Edit Distance (Levenshtein Distance)

Given two strings `s1` and `s2`, find the minimum number of operations
(**Insert**, **Remove**, **Replace**) required to convert `s1` into `s2`.
All operations have equal cost.

We solve this using **Bottom-Up Dynamic Programming (Tabulation)**:
- Time Complexity: `O(m * n)`
- Space Complexity: `O(m * n)`

Where `m = len(s1)` and `n = len(s2)`.

````python
# Import required libraries
import pandas as pd  # used only to nicely display the DP table
````

## Step 1: Function to build the DP table

- Create a table `dp` of size `(m+1) x (n+1)`.
- Initialize first row `dp[0][j] = j` and first column `dp[i][0] = i`
  (base case: converting an empty string to a string of length `j`/`i`
  requires `j`/`i` insertions).
- Fill the rest of the table using:
  - If `s1[i-1] == s2[j-1]`: `dp[i][j] = dp[i-1][j-1]`
  - Else: `dp[i][j] = 1 + min(dp[i][j-1], dp[i-1][j], dp[i-1][j-1])`

````python
def min_edit_distance(s1, s2, return_table=False):
    """
    Compute the minimum edit distance between s1 and s2
    using bottom-up dynamic programming (tabulation).

    Parameters
    ----------
    s1, s2 : str
        Input strings.
    return_table : bool
        If True, also return the full DP table.

    Returns
    -------
    int  (and list[list[int]] if return_table=True)
    """
    m, n = len(s1), len(s2)

    # Step 1: Create (m+1) x (n+1) table initialized with zeros
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]

    # Step 2: Initialize base cases (first row and first column)
    for i in range(m + 1):
        dp[i][0] = i          # converting s1[0..i] to empty string -> i deletions
    for j in range(n + 1):
        dp[0][j] = j          # converting empty string to s2[0..j] -> j insertions

    # Step 3: Fill the rest of the table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(
                    dp[i][j - 1],      # Insert
                    dp[i - 1][j],      # Remove
                    dp[i - 1][j - 1]   # Replace
                )

    if return_table:
        return dp[m][n], dp
    return dp[m][n]
````

## Step 2: Helper function to visualize the DP table

This displays the table exactly like the worked example, with
row/column labels corresponding to the characters of `s1` and `s2`.

````python
def show_dp_table(s1, s2, dp):
    """Pretty-print the DP table using pandas, with s1/s2 characters as labels."""
    row_labels = [""] + list(s1)
    col_labels = [""] + list(s2)
    df = pd.DataFrame(dp, index=row_labels, columns=col_labels)
    return df
````

## Step 3: Test with the example from the reference

```
s1 = "abcd"
s2 = "bcfe"
```

Expected final DP table (bottom-right corner = answer):

| | 0 | b | c | f | e |
|---|---|---|---|---|---|
|0| 0 | 1 | 2 | 3 | 4 |
|a| 1 | 1 | 2 | 3 | 4 |
|b| 2 | 1 | 2 | 3 | 4 |
|c| 3 | 2 | 1 | 2 | 3 |
|d| 4 | 3 | 2 | 2 | 3 |

So the minimum edit distance should be **3**.

````python
s1 = "abcd"
s2 = "bcfe"

distance, dp_table = min_edit_distance(s1, s2, return_table=True)

print(f"s1 = '{s1}'")
print(f"s2 = '{s2}'")
print(f"Minimum Edit Distance = {distance}")
````

````python
# Display the full DP table
show_dp_table(s1, s2, dp_table)
````

## Step 4: Try it with your own strings

````python
# Try with different strings
s1 = "kitten"
s2 = "sitting"

distance, dp_table = min_edit_distance(s1, s2, return_table=True)

print(f"s1 = '{s1}'")
print(f"s2 = '{s2}'")
print(f"Minimum Edit Distance = {distance}")

show_dp_table(s1, s2, dp_table)
````

## Complexity

- **Time Complexity:** `O(m * n)` — we fill every cell of the table once.
- **Space Complexity:** `O(m * n)` — for the full table (can be optimized to `O(min(m, n))` if only the distance, not the table, is needed).
